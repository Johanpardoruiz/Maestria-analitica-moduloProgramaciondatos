```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
x = np.array([1, 2, 3, 4, 5]) 
y = np.array([1, 3, 6, 7, 10]) 

# cear la matriz de diseño X con una columna de 1's para el intercepto
X = np.vstack([np.ones_like(x),x]).T

######## CALCULO DE LOS BETAS
# Transpuesta de X
X_transpuesta = X.T

# (X^T X)
XtX = np.dot(X_transpuesta, X)

# (X^T X)^-1
XtX_inversa = np.linalg.inv(XtX)

# (X^T y)
Xty = np.dot(X_transpuesta, y)

# ((X^T X)^-1)((X^T y))
beta = np.dot(XtX_inversa, Xty)


#Los coeficientes beta represetan [b,m] para la ecuacion y=m*x+b
intercepto = beta[0]  # b (intercepto)
pendiente = beta[1]   # m (pendiente)


print(f"Coeficiente (intercepto) b: {intercepto}")
print(f"pendiente m: {pendiente}")


#predicciones con la recta de regresión
y_pred = np.dot(X, beta) #Y=m*X+b 

#grafica datos originales(puntos)
plt.scatter(x,y,color="blue", label="Datos", marker="o")


# Graficar la línea de regresión
plt.plot(x, y_pred, color='red', label=f'Regresión Lineal: y = {pendiente:.2f} x + {intercepto:.2f}', linewidth=2)

# Etiquetas y título
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.title('Regresión Lineal: Horas de Estudio vs Calificación')

# Mostrar leyenda
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()




```
